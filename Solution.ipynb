```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("F:\\Contests\\NebulaNet\\nebulanet\\train.csv")
df_test = pd.read_csv("F:\\Contests\\NebulaNet\\nebulanet\\test.csv")

df_test
```

```python
df_train.dtypes
```

```python
from sklearn import preprocessing

labelencode = preprocessing.LabelEncoder()
#labeltransform = ColumnTransformer([("labelencode", labelencode, ["Star type"])], remainder= "passthrough")
Star_type_encoded = labelencode.fit_transform(df_train["Star type"])
Star_type_encoded
```

```python
df_train["Star type"] = Star_type_encoded
df_train
```

```python
df_test["Star type"] = 0
df_test
```

```python
df = pd.concat([df_train, df_test])
df
```

```python
x = df.drop(columns="Star type", axis = 1)
```

```python
from sklearn.compose import ColumnTransformer

cf = ["Spectral Class", "Star color"]
encode = preprocessing.OneHotEncoder(sparse_output= False)

transform = ColumnTransformer([("encode", encode, cf)], remainder= "passthrough")
transformed_x = transform.fit_transform(x)
#transformed_x_test = transform.fit_transform(df_test)
#transformed_x
```

```python
max_abs_scale = preprocessing.MaxAbsScaler()
x_train_abs_scale= max_abs_scale.fit_transform(transformed_x)
#x_test_abs_scale= max_abs_scale.fit_transform(transformed_x_test)
#x_train_abs_scale
```

```python
x1 = pd.DataFrame(x_train_abs_scale)
x_train = x1[:-24]
x_test = x1[-24:]
y_train = df["Star type"][:-24]
#x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
#from sklearn.model_selection import train_test_split
np.random.seed(42)

#x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, test_size= 0.2)

model = RandomForestClassifier()
model.fit(x_train, y_train)
y_test = model.predict(x_test)
y_test

```

```python
y_label_pred= labelencode.inverse_transform(y_test)

```

```python
ID = np.arange(1, 25, 1)
ID
```

```python
solution = pd.DataFrame([ID, y_label_pred]).T
solution.rename(columns= {0: "ID", 1: "Star type"}, inplace=True)
solution
```

```python
solution.to_csv(r"F:\Contests\NebulaNet\nebulanet\Solution.csv", index= False)
```
